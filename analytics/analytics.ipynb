```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams["figure.figsize"] = (24,4)
```

```python

with open('stats_raw.csv', 'r') as f:
    stats = pd.read_csv(f)
#stats
```

```python
# Calculate runtime compared to moves considered
stats['timePerMove'] = stats['runtime'] / stats['movesConsid']

# Plot this data
colors = ['#ffafcc', '#a2d2ff']
fig, ax1 = plt.subplots() 
  
ax1.set_xlabel('Round Number') 
ax1.set_ylabel('Moves considered', color = colors[0]) 
plt.plot(stats['roundID'], stats['movesConsid'], color = colors[0], lw=0.5)
ax1.tick_params(axis ='y', labelcolor = colors[0]) 
  
# Adding Twin Axes
ax2 = ax1.twinx() 
  
ax2.set_ylabel('Consideration time per move', color = colors[1]) 
plt.plot(stats['roundID'], stats['timePerMove'], color = colors[1], lw=1)
ax2.tick_params(axis ='y', labelcolor = colors[1]) 
 
plt.savefig('plots/timePerMove.png', dpi=300)
```

```python
# Compare moves considered and moves executed

# Plot this data
colors = ['#ffafcc', '#a2d2ff']
fig, ax1 = plt.subplots() 
  
ax1.set_xlabel('Round Number') 
ax1.set_ylabel('Moves considered', color = colors[0]) 
plt.plot(stats['roundID'], stats['movesConsid'], color = colors[0], lw=1)
ax1.tick_params(axis ='y', labelcolor = colors[0]) 
  
# Adding Twin Axes
ax2 = ax1.twinx() 
  
ax2.set_ylabel('Moves actually executed', color = colors[1]) 
plt.plot(stats['roundID'], stats['movesExec'], color = colors[1], lw=0.5)
ax2.tick_params(axis ='y', labelcolor = colors[1]) 
 
plt.savefig('plots/movesRatio.png', dpi=800)
```

```python
# Compare move efficiency
stats['moveEfficiency'] = stats['movesConsid'] / stats['movesExec']

# Plot this data
colors = ['#ffafcc', '#a2d2ff']

plt.plot(stats['roundID'], stats['moveEfficiency'], color = colors[0], lw=1)
plt.ylabel('Moves considered / moves executed')
plt.xlabel('Game Round #')
plt.title('Efficiency factor of move considerations')
 
plt.savefig('plots/movesEfficiency.png', dpi=800)
```
